#### Q6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def calculate_median(data):
    
    #  sorting in ascending order
    sorted_data = sorted(data)
    
    n = len(sorted_data)
    if n % 2 == 1:
        # If the number of data points is odd, return the middle value
        median_index = n // 2
        median = sorted_data[median_index]
    else:
        # If the number of data points is even, return the average of the two middle values
        right_middle_index = n // 2
        left_middle_index = right_middle_index - 1
        median = (sorted_data[left_middle_index] + sorted_data[right_middle_index]) / 2
        
    return median
```

##### now we can calculate median value of any column - just put the column name in the place of 'total_bedrooms'

```python
column_name = 'total_bedrooms'
column_data = data[column_name]

# Calling the function to calculate the median
median_value = calculate_median(column_data)

# Output the result
print("Median of column '{}': {}".format(column_name, median_value))
```

#### Q10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
# function to categorize the total bedrooms size
def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif total_bedrooms >= 11 and total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

#  created the 'total_bedroom_size' column
data['total_bedroom_size'] = data['total_bedrooms'].apply(categorize_bedroom_size)

# Display of DataFrame with the new 'total_bedroom_size' column
print(data)
```
